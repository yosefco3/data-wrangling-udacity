## Clean

1. in the "image predictions" dataset: 
 * if the sysytem didn't recognize a dog at all, we will remove this row. 
 * if there was a dog recognization, we would take the one with the highest probability.

#### first lets unite our datasets:
twitter_archive = twitter_archive.merge(df_retweet_favorite, how='inner',left_on='tweet_id', right_on='id')

```python

image_predictions.head(2)
```

```python
# remove rows that there isn't a dog recognition at all:
image_predictions = image_predictions[image_predictions["p1_dog"] |  image_predictions["p2_dog"] | \
                                      image_predictions["p3_dog"]]

# take the dog with the highest probability:
def find_dog(row):
    if row["p1_dog"]:
        return row["p1"]
    elif row["p2_dog"]:
        return row["p2"]
    elif row["p3_dog"]:
        return row["p3"]
    else:
        return ""
    
image_predictions["dog_kind"]= image_predictions.apply(find_dog,axis = 1)
```

```python
# test:
image_predictions[image_predictions["p1_dog"]==False].head()
```

```python
# test - id with 3 false - need to return 0:
(image_predictions["tweet_id"]==668643542311546000).sum()
```

```python
final_image_prediction = image_predictions[["tweet_id","dog_kind"]]
final_image_prediction
```

```python
# now we will inner join final_image_prediction with the archive:
archive = twitter_archive_extract.merge(final_image_prediction, how='inner', on='tweet_id')
archive.head()
```

2. we will remove other columns we dont need:

```python
archive.drop(columns=["text","expanded_urls","name"],inplace = True)
archive.head()
```

3. change "timestamp" from object to datetime:

```python
archive["timestamp"] = pd.to_datetime(archive["timestamp"])
archive.info()
```

4. in some columns there are 0 values in favorite_count and thousands in retweet_count, it's not logical! so we would replace them with the mean\median value


```python
med = archive["favorite_count"].median()

def replace(x):
    if x==0:
        return med
    else:
        return x

archive["favorite_count"] = archive["favorite_count"].apply(replace)
```

```python
# test:
(archive["favorite_count"]==0).sum()
```

5. the normal rating_numerator is like 10-20 , but there are nominators of hundreds ...and the max value in thousands. so if the rating_numerator is above 15 we will replace it with 15 (high value, but not too much)

```python

def replace(x):
    if x>15:
        return 15
    else:
        return x
archive["rating_numerator"] = archive["rating_numerator"].apply(replace)
```

```python
# test :
archive["rating_numerator"].max()
```

6. the mean of the rating_denominator isn't 10

```python
archive["rating_denominator"] = 10
```

```python
# test:
archive["rating_denominator"].mean()
```

7. i was prefer all the "doggo,floofer,pupper,puppo" in one column, so i could do value counts easily.

```python
archive.head(2)
```

```python
# make all 4 columns of "dog stage" to one, and replace the "None" with numpy NaN:
def staging(row):
    if row["doggo"] !="None":
        return "doggo"
    elif row["floofer"]!= "None":
        return "floofer"
    elif row["pupper"]!="None":
        return "pupper"
    elif row["puppo"]!="None":
        return "puppo"
    else:
        return np.nan
    
    
archive["stage"] = archive.apply(staging,axis =1)
# test:
archive.head(20)
# we see the staging in the new column "stage" we added:
```

```python
# test:
archive.drop(columns=["doggo","floofer","pupper","puppo"],inplace=True)
archive.head()
```

```python
## This is the final and clean version of the archive, so we will save it to csv:
archive.to_csv("twitter_archive_master.csv",index=False)
```

## Explore and visualizations:

* what are the most common dog kinds ?

```python
most_popular_dogs = twitter_archive_master["dog_kind"].value_counts()[:10]
most_popular_dogs = np.around(most_popular_dogs/most_popular_dogs.sum(), decimals=2)*100
most_popular_dogs
```

```python
#  a function to create bar plot:
def create_plot(x,y,
                plot_color,
                xlab, ylab,
                lab_size,
                title,
                title_size,
                lab_and_title_color,
                xtick_rotation,
                tick_size,tick_color,fig_size_x=15,fig_size_y=10,plt_kind='bar'):
    fig = plt.figure(figsize = (fig_size_x, fig_size_y)) 
      # creating the bar plot 
    if plt_kind=='bar':
        plt.bar(x,y, color =plot_color);
    plt.xlabel(xlab,fontsize = lab_size,color = lab_and_title_color) 
    plt.ylabel(ylab,fontsize = lab_size,color = lab_and_title_color) 
    plt.title(title,fontsize=title_size,color = lab_and_title_color) 
    plt.xticks(rotation = xtick_rotation)
    plt.tick_params(labelsize=tick_size,labelcolor=tick_color)
```

```python
create_plot(x=most_popular_dogs.index,y=most_popular_dogs.values,
                plot_color='#42c8f5',
                xlab ="Dog breed", ylab = "Dogs precentage", 
                lab_size=20,
                title="The most popular dog breeds in the archive",
                title_size =20,
                lab_and_title_color='green',
                xtick_rotation =70,
                tick_size=18,
                tick_color='red')
```

* what are the most favorites dog kinds ?

```python
most_favorites_dogs = np.around(twitter_archive_master.groupby("dog_kind").mean()["favorite_count"].\
sort_values(ascending=False)[:10]/1000,2)
most_favorites_dogs
```

```python
create_plot(x=most_favorites_dogs.index,y=most_favorites_dogs.values,
                plot_color='navy',
                xlab ="Dog breed", ylab = "Number of favorites (Thousands)", 
                lab_size=20,
                title="The most favorated dog breeds in the archive",
                title_size =20,
                lab_and_title_color='green',
                xtick_rotation =70,
                tick_size=18,
                tick_color='red')
```
